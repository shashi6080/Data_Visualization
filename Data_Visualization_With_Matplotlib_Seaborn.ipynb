# Data Visualization Using Matplotlib and Seaborn

## 1. Plot two variables lists

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 49, 46]
plt.plot(x, y)
```

## 2. Plot, add line color

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 49, 46]
plt.plot(x, y, color='green')
```

## 3. Plot, add line width

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 49, 46]
plt.plot(x, y, color='blue', linewidth=5)
```

Note :<br>
<b>Linestyle	          ----> Description</b><br>
'-' or 'solid'	      ----> solid line<br>
'--' or 'dashed'	  ----> dashed line<br>
'-.' or 'dashdot'	  ----> dash-dotted line<br>
':' or 'dotted'	      ----> dotted line<br>
'None' or ' ' or ''	  ----> draw nothing<br>

## 4. Plot, line style

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 49, 46]
plt.plot(x, y, color='red', linewidth=2, linestyle='--')
```

## 5. Plot, add xlabel and ylabel

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 49, 46]
plt.xlabel('Weekday')
plt.ylabel('Temparature')
plt.plot(x, y, color='black', linewidth=3, linestyle=':')
```

## 6. Plot two lists, add title

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 49, 46]
plt.xlabel('Weekday')
plt.ylabel('Temparature')
plt.title('Predicted Weather')
plt.plot(x, y, color='gray', linewidth=3, linestyle='-.')
```

<b>Format Strings:</b><br>
    A format string consists of a part for color, marker and line:<br>
        fmt = '[marker][line][color]'<br>

## 7. Plot, add marker

## 9. Plot, with marker size

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 49, 46]
plt.xlabel('Weekday')
plt.ylabel('Temparature')
plt.title('Predicted Weather')
plt.plot(x, y, marker='+', linestyle='', color='#888888', markersize=20)
```

## 10. Plot, with alpha

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 49, 46]
plt.xlabel('Weekday')
plt.ylabel('Temparature')
plt.title('Predicted Weather')
plt.plot(x, y, marker='*', linestyle='--', color='#668866', markersize=20, alpha=0.5) # color value is hexa decimal
```

## 18. Bar Chart with Legend

```python
company = ['Google', 'Amazon', 'Microsoft', 'Facebook']
revenue = [90, 130, 89, 27]
xpos = np.arange(len(company))

plt.xticks(ypos, company) # this function converts integer with string labels
plt.ylabel('revenue(billion)')
plt.title('Stocks')
plt.bar(xpos, revenue, label='Revenue') # label value adds to the legend
plt.legend()
```

## 19. Plot Two Bar Charts 

```python
company = ['Google', 'Amazon', 'Microsoft', 'Facebook']
revenue = [90, 130, 89, 27]
profit = [20, 30, 15, 10]
xpos = np.arange(len(company))

plt.xticks(ypos, company) # this function converts integer with string labels
plt.ylabel('revenue(billion)')
plt.title('Stocks')
plt.bar(xpos, revenue, label='Revenue')# label value adds to the legend
plt.bar(xpos, profit, label='Profit')
plt.legend()
```

<b>Note:</b><br>
    * If we are plotting two bar charts then their position is same.<br>
    * We need to explicitly alter position to view charts side by side.<br>

## 20. Plot Two Bar Charts Side-by-Side

```python
company = ['Google', 'Amazon', 'Microsoft', 'Facebook']
revenue = [90, 130, 89, 27]
profit = [20, 30, 15, 10]
xpos = np.arange(len(company))

plt.xticks(ypos, company) # this function converts integer with string labels
plt.ylabel('revenue(billion)')
plt.title('Stocks')
plt.bar(xpos-0.2, revenue, width=0.4, label='Revenue') # altered x position and set width
plt.bar(xpos+0.2, profit, width=0.4, label='Profit')  # altered x position and set width
plt.legend()
```

## 21. Plot Horizontal Bar Charts

```python
company = ['Google', 'Amazon', 'Microsoft', 'Facebook']
revenue = [90, 130, 89, 27]
profit = [20, 30, 15, 10]
ypos = np.arange(len(company))

plt.yticks(ypos, company) # yticks function ticks y-axis labels
plt.ylabel('revenue(billion)')
plt.title('Stocks')
plt.barh(ypos-0.2, revenue, label='Revenue') # altered y position and set width
plt.barh(ypos+0.2, profit, label='Profit')  # altered y position and set width
plt.legend()
```

## 22. Plot Histograms

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 10, 48, 92, 49, 87, 62, 30, 10, 39, 48, 77]
plt.hist(blood_sugar, rwidth=0.9) # max value of rwidth is 1
```

## 23. Plot Histograms, with Bins and Color

```python
blood_sugar = [113, 85, 10, 48, 92, 49, 87, 62, 30, 10, 39, 48, 77]
plt.hist(blood_sugar, bins=5, rwidth=0.9, color='green') # bins -> no. of equal width bins in tha range
```

## 24. Plot Two Histograms

```python
blood_sugar_men = [113, 85, 10, 48, 92, 49, 87, 62, 30, 10]
blood_sugar_women = [90, 105, 20, 59, 30, 80, 55, 90, 70, 60]
plt.xlabel('Sugar Range')
plt.ylabel('Total no. of Patients')
plt.title('Blood Sugar Analyis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=5, rwidth=0.9, color=['green','orange'],
         label=['Men', 'Women']) # bins -> no. of equal width bins in tha range
plt.legend()
```

## 25. Plot Two Histograms Horizontal

```python
blood_sugar_men = [113, 85, 10, 48, 92, 49, 87, 62, 30, 10]
blood_sugar_women = [90, 105, 20, 59, 30, 80, 55, 90, 70, 60]
plt.ylabel('Sugar Range')
plt.xlabel('Total no. of Patients')
plt.title('Blood Sugar Analyis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=5, rwidth=0.9, color=['green','orange'],
         label=['Men', 'Women'], orientation='horizontal') # bins -> no. of equal width bins in tha range
plt.legend()
```

## 26. Plot Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ['Home Rent', 'Food', 'Phone Bill', 'Car', 'Others']
```

```python
plt.pie(exp_vals, labels=exp_labels)
plt.show()  # this is optional and recommended to avois unnecessary print
```

## 26. Plot Pie Chart, With Specified Radius

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ['Home Rent', 'Food', 'Phone Bill', 'Car', 'Others']

plt.pie(exp_vals, labels=exp_labels, radius=1.5) # default value of Radius is 1
plt.show()  # this is optional and recommended to avois unnecessary print
```

## 27. Plot Pie Chart, With Percentage

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ['Home Rent', 'Food', 'Phone Bill', 'Car', 'Others']

plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.2f%%') # autopct value is with 2 decimal points
plt.show()  # this is optional and recommended to avois unnecessary print
```

## 28. Plot Pie Chart, With Explode

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ['Home Rent', 'Food', 'Phone Bill', 'Car', 'Others']

plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.2f%%', shadow=True, explode=[0, 0.2, 0.2, 0, 0]) # autopct value is with 2 decimal points
plt.show()  # this is optional and recommended to avois unnecessary print
```

## 29. Plot Pie Chart, With Start Angle

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ['Home Rent', 'Food', 'Phone Bill', 'Car', 'Others']

plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.2f%%', shadow=True, 
        explode=[0, 0.2, 0.2, 0, 0], startangle=180) # autopct value is with 2 decimal points
plt.show()  # this is optional and recommended to avois unnecessary print
```

## 30. Save Image to File

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ['Home Rent', 'Food', 'Phone Bill', 'Car', 'Others']

plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.2f%%', 
        explode=[0, 0.2, 0.2, 0, 0], startangle=180) # autopct value is with 2 decimal points
plt.savefig('piechart.png', bbox_inches='tight', pad_inches=2, transparent=True)
```
