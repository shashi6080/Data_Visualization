<b>Introduction to pyplot</b>
<br>
matplotlib.pyplot is a collection of command style functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

In matplotlib.pyplot various states are preserved across function calls, so that it keeps track of things like the current figure and plotting area, and the plotting functions are directed to the current axes (please note that "axes" here and in most places in the documentation refers to the axes part of a figure and not the strict mathematical term for more than one axis).

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Example 1 : Line plot
plt.plot([2, 5, 6, 4])
plt.ylabel('Numbers')  #xlabel name
plt.xlabel('Indices') #ylabel name
plt.title('MyPlot') #title of the picture
plt.show() #show picture

#note : when a List is given as input, the values are plotted into y-axis and indices being x-axis values.
```

```python
#Example 2 : Line plot
plt.plot([1, 2, 3, 4],[1, 4, 8, 16])
plt.xlabel('numbers')
plt.ylabel('squares')
plt.grid()  #for grid like figure
plt.show()

#note: we have provided explicit x-axis and y-axis values. Hence those will be taken for plotting.
#note: Grid is more readable
```

```python
#Example 3 : Line plot
t = np.arange(0.0, 2.0, 0.01) #numpy range function
s = 1 + np.sin(2*np.pi*t) #an equation 
plt.plot(t, s)

plt.xlabel('time')
plt.ylabel('voltage')
plt.title('time voltage plotting')
plt.grid(True) #grid required
plt.savefig('test.png') #to save plotted figure as file
plt.show()

#note : Here we have explicitly provided x-axis and y-axis values. And these values will be plotted.
```

```python
#Example 4 : Dot plot(Red)
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') #r->red, 0->dotted plot
plt.xlabel('numbers')
plt.ylabel('squares')
plt.title('squares plotting')
plt.grid()
```

```python
#Example 5 : Dot plot(Green)
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'go') #g->green, 0->dotted plot
plt.xlabel('numbers')
plt.ylabel('squares')
plt.title('squares plotting')
plt.grid()
```

```python
#Example 6 : Multiple plots
t = np.arange(0., 5., 0.2)

plt.plot(t, t**2, 'b--', label='^2') #'b--' -> blue dashes
plt.plot(t, t**2.2, 'rs', label='^2.2') #'rs' -> red squares
plt.plot(t, t**2.5, 'g^', label='^2.5')#'g^' -> green triangles
plt.grid()
plt.legend() # add legend based on line labels
plt.show()
```

## Controlling Line Properties

```python
#Example 7 : Line plot with Line width
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y, linewidth=7.0)
plt.show()
```

```python
#Example 8 : Line plot with line width and red color
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y, 'r', linewidth=5.0)
plt.show()
```

```python
#Example 9 : Two plots using Set Properties
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
x2 = [1, 2, 3, 4]
y2 = [2, 4, 6, 8]

lines = plt.plot(x1, y1, x2, y2) #which return two lines lines[0] and lines[1]

# using keyword arguments
plt.setp(lines[0], color='r', linewidth='2.0') #set properties for line 0

#using MATLAB style string valuespairs #set properties for line 1
plt.setp(lines[1], 'color', 'g', 'linewidth', '2.0')
plt.grid()
plt.show()
```

## Working with multiple figures and axes

```python
#Example 10 : One figure with multiple sub-figures
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)

plt.subplot(211) #211 -> no. of rows(2), no. of columns(1) and first figure(1)
plt.grid()
plt.plot(t1, f(t1), 'b-')

plt.subplot(212) #212 -> no. of rows(2), no. of columns(1) and second figure(2)
plt.grid()
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
#Example 11 : Multiple figures with multiple sub-figures
plt.figure(1)  #the first figure
plt.subplot(211)  #the first subplot in the first figure
plt.plot([1, 2, 3])  
plt.subplot(212)  #the second subplot in the first figure
plt.plot([4, 6, 12])

plt.figure(2)  #the second figure
plt.plot([2, 5, 10])  #creates a subplot(111) by default

plt.figure(1)  #figure 1 current. subplot(212) still current
plt.subplot(211)  # make subplot(211) in figure1 as current
plt.title('Easy as 1, 2, 3')  #subplot 211 title
plt.show()
```
